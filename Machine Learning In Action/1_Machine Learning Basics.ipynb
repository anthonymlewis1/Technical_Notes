# Machine Learning Basics

## What is Machine Learning?
* Machine learning turns data into information
* Machine learning is used in any field that needs to interpret/act on data
* The solutions to machine learning are not deterministic though
* Why is this important?
    * We will only have more and more data as time progresses
    * Society has moved from manual labor to knowledge work

## Key Terminology
* Expert Systems: Replacing specialized occupations with a computer
* Features: Measurements of a dataset
* Instance: Rows of data
* Classification: Identifying which of a set of categories an observation belongs to
* Training Set: Examples to train an ML model
* Target Variable: What we're trying to predict with the model
    * Classification: Nominal
    * Regression: Continuous
    * Known within a training set
* Test Set: Validation set to check how accurate the algorithm is.
* Knowledge Representation: What has the machine learned?

## Key Tasks of Machine Learning
* Classification -> Predict what class an instance of data should fall into
* Regression -> Prediction of a numeric value
* Supervised Learning: When you tell the algorithm what to predict
* Unsupervised Learning: No label/target variable is given to predict
* Clustering: Grouping similar items together
* Density Estimation: Statistical values that describe data used in unsupervised learning
* Supervised Learning Algorithms:
    * k-Nearest Neighbors
    * Naive Bayes
    * Support Vector Machines
    * Decision Trees
* Unsupervised Learning Algorithms:
    * K-Means
    * DBSCAN

## Choosing the Right Algorithm
* Are you trying to predict a target value?
    * Yes: Supervised learning
        * What's your target value?
            * Discrete: Classification
            * Continuous: Regression
    * No: Unsupervised learning
        * Are you trying to fit your data into similar groups?
            * Yes: Clustering
        * Are you trying to see how strong the fit is in each group?
            * Yes: Density Estimation

## Steps in Developing a Machine Learning Application
* Collect data
* Prepare the input data
* Analyze the input data
* Train the Algorithm
* Test the Algorithm
* Use it
