# 1: Getting Started with Terraform

## What Is Terraform:
* Deployment technology for anyone who wants to provision and manage their infrastructure as code
* Basic Principle: Write Human-readable config code to define IAC
* Infrastructure as Code (IAC):
    * Infrastructure through machine-readable definition files
    * Automated processes!
* Provisioning:
    * Act of deploying infrastructure
    * This is NOT Configuration Management
        * CM typically deals with application delivery typically on VMs

## Benefits of Terraform:
    * Provisioning Tool:
        * Deploys infrastructure, not just apps
        * Immutable Infrastructure: Treat infra as a disposable commodity
        * Mutable Infrastructure: Make updates / patches to infra
    * Easy to Use:
        * I'm dumb so I need things dumbed down for me
        * Written in HashiCorp Configuration Language (HCL)
            * Compatible with JSON
        * All config based 
    * Open Source:
        * Free!!
    * Declarative:
        * Say what you want
        * As opposed to functional programming (How to do what you want)
    * Cloud-Agnostic: Same tools for different platform
    * Expressive and Extendable: Easy to add new functionality via new code
    
### Writing in Terraform:

1) Write Terraform Config
2) Configure AWS providor
3) Initialize Terraform
4) Deploy terraform
5) Clean up terraform

#### Setting up the Configs
```
# main.tf
# Declare an AWS EC2 instance with an Ubuntu AMI (type) and a name tag

provider "aws" {
region = "us-west-2"
}

data "aws_ami" "ubuntu" {
  most_recent = true
    filter {
    name = "name"
      values = ["ubuntu/images/hvm-ssd/ubuntu-focal-20.04-amd64-server-*"]
    }
    owners = ["099720109477"] # AWS Account ID


resource "aws_instance" "helloworld" {
  ami           = "ami-09dd2e08d601bff67"
  instance_type = "t2.micro"
  tags = {
    Name = "HelloWorld"
  }
}
```
           
* Provider: (provider "aws"{ ... })
    * Responsible for handling API interactions and exposing resources to Terraform
    * Many providers for different systems are noted in the Terraform Registry
        * Ex: google: gcp, azurerm: azure
    * Providers only have inputs
        * Configruation Argument: Input
        
* Data Source:
    * Elements that allow you to fetch data at runtime / perform computations
    * TF can only read no control
    * In the example, it queries the most recent Uvuntu AMI published to the Registry

* Resource: (resource "aws_instance" "helloworld" { ... })
    * HCL defined element for specifcying infrastructure building blocks
        * Ex: Used for VMs, load balancers, etc.
    * Each resource has an input and an output
        * Argument: Input
        * Attributes: Output
            * Computed / Calculated Attributes: Only available after creation

           
            
####  Initializing,  Applying, and Destroying Terraform
```
terraform init # Download binaries and lock the config
terraform apply # Deploy and run through plan
terraform destroy # Destroys created infrastructure
```
* *terraform apply* creates an execution plan that outlines the actions to achieve the desired state
